```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, LabelEncoder, FunctionTransformer, OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split before preprocessing to avoid data leakage
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop('Depression', axis=1), df['Depression'], test_size=0.2, random_state=42)
```

```python
# Encode target separately
label_enc = LabelEncoder()
y_train = label_enc.fit_transform(y_train)
y_test = label_enc.transform(y_test)
```

```python
num_cols = ['Age', 'Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress']
binary_cols = ['Gender', 'Family History of Mental Illness']
ordinal_cols = ['Sleep Duration', 'Dietary Habits']
```

```python
# Custom orders for Ordinal Encoding
sleep_order = [
    "Less than 5 hours",
    "5-6 hours",
    "7-8 hours",
    "More than 8 hours"
]
diet_order = ["Unhealthy", "Moderate", "Healthy"]
```

```python
import numpy as np
```

```python
# Binary encoder function (mapping Yes and "Male" to 1 and alternative to 0 using Function Transformer)
def binary_map(X):
    return np.where((X == 'Yes') | (X == 'Male'), 1, 0).astype(int)

binary_transformer = FunctionTransformer(binary_map, validate=False)
```

```python
# Build ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), num_cols),
        ('binary', binary_transformer, binary_cols),
        ('ordinal', OrdinalEncoder(categories=[sleep_order, diet_order]), ordinal_cols)
    ],
    remainder='drop'
)
```

```python
# Fit-transform
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, classification_report, confusion_matrix,
                             ConfusionMatrixDisplay)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,      # number of trees
    max_depth=None,        # try integers to regularize
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_transformed, y_train)
```

```python
y_pred_rf = rf.predict(X_test_transformed)
# For ROC-AUC / probability-based metrics (binary):
y_proba = rf.predict_proba(X_test_transformed)[:, 1]   # probability for positive class
```

```python
acc_rf   = accuracy_score(y_test, y_pred_rf)
prec_rf  = precision_score(y_test, y_pred_rf, average='binary')   # use 'weighted' for multiclass
rec_rf  = recall_score(y_test, y_pred_rf, average='binary')
f1_rf   = f1_score(y_test, y_pred_rf, average='binary')

# ROC AUC (binary):
roc_auc_rf = roc_auc_score(y_test, y_proba)

print("accuracy:", acc_rf)
print("precision:", prec_rf)
print("recall:", rec_rf)
print("f1:", f1_rf)
print("roc_auc:", roc_auc_rf)


# Store metrics in a dictionary
rf_results = {
    "Model": "Random Forest",
    "Accuracy": acc_rf,
    "Precision": prec_rf,
    "Recall": rec_rf,
    "F1 Score": f1_rf,
    "ROC AUC": roc_auc_rf
}



```

```python
print(classification_report(y_test, y_pred_rf))   # precision, recall, f1 per class
cm = confusion_matrix(y_test, y_pred_rf)
ConfusionMatrixDisplay(cm).plot()
plt.show()
```

```python
metrics_list = []
metrics_list.append(rf_results)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, BaggingClassifier
```

```python
svc = SVC(probability=True, random_state=42)
knn = KNeighborsClassifier(n_neighbors=6, weights='distance')
rf = RandomForestClassifier(random_state=42)
nb = GaussianNB()
gb= GradientBoostingClassifier(n_estimators=100, random_state=42)
```

```python
bagging_svc = BaggingClassifier(estimator=svc, n_estimators=10, random_state=42)
bagging_knn = BaggingClassifier(estimator=knn, n_estimators=10, random_state=42)
bagging_rf = BaggingClassifier(estimator=rf, n_estimators=10, random_state=42)
bagging_nb = BaggingClassifier(estimator=nb, n_estimators=10, random_state=42)
bagging_gb = BaggingClassifier(estimator=gb, n_estimators=10, random_state=42)
```

```python
bagging_models = {
    "Bagging_SVC": bagging_svc,
    "Bagging_KNN": bagging_knn,
    "Bagging_RF": bagging_rf,
    "Bagging_NB": bagging_nb,
    "Bagging_GB": bagging_gb
}
```

```python
for name, model in bagging_models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)

    metrics_list.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="macro"),
        "Recall": recall_score(y_test, y_pred, average="macro"),
        "F1 Score": f1_score(y_test, y_pred, average="macro"),
        "ROC AUC": roc_auc_score(y_test, y_pred)
    })
```

```python
df_metrics=pd.DataFrame(metrics_list)
```
